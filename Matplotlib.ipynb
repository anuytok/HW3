```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 8), tight_layout=True)

x_ticks = ['PBP1B', 'PBP1B\n+ BacA', 'PBP1B\n + PgpB', 'PBP1B\n + PgpB*']

x1_values = [100, 110, 119, 124]
x1_errors = [[0, 0, 0, 0], [0, 5, 10, 12]]
axs[0].set_title('B', loc='left', pad=20,
                 fontdict={'fontsize': 20, 'horizontalalignment': 'left'})
axs[0].bar(x_ticks, x1_values, width=0.6,
           yerr=x1_errors, error_kw={'capsize': 3.0},
           color=['#BCBBBA', '#9B9998', '#5C5B5B', '#343333'],
           edgecolor='black')
axs[0].set_ylabel('% PG products relative PBP1B alone',
                  fontdict={'fontsize': 14})
axs[0].set_yticks([60, 80, 100, 120, 140])
axs[0].set_ylim(ymin=60)
axs[0].tick_params(labelsize='large')
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
y = 1.1*max(x1_values[0], x1_values[2])
props = {'connectionstyle':'bar','arrowstyle':'-',
         'shrinkA':35,'shrinkB':35,'linewidth':2}
axs[0].annotate("$\it{p}$ = 0.011", xy=(0.6, y+10),
                annotation_clip=False, size=14)
axs[0].annotate('', xy=(0, y-12), xytext=(2, y-12), arrowprops=props)

x2_values = [42, 46, 48, 49]
x2_errors = [[0, 0, 0, 0], [8, 5, 3, 4]]
axs[1].set_title('C', loc='left', pad=20,
                 fontdict={'fontsize': 20, 'horizontalalignment': 'left'})
axs[1].bar(x_ticks, x2_values, width=0.6, yerr=x2_errors,
           error_kw={'capsize': 3.0},
           color=['#BCBBBA', '#9B9998', '#5C5B5B', '#343333'],
           edgecolor='black')
axs[1].set_ylabel('% of peptides in crosslinks', fontdict={'fontsize': 14})
axs[1].set_yticks([0, 10, 20, 30, 40, 50, 60])
axs[1].tick_params(labelsize='large')
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)


fig.savefig('result.png')
plt.show()
```
